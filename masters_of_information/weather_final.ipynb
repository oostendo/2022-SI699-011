```python

\
all_csvs = []

for root, dirs, files in os.walk("."):
    for file in files:
        if file.endswith('.csv') and (file.startswith('2') or file.startswith('1')):
            
            all_csvs.append(file)
            
```

```python
all_csvs
```

```python
df_all = pd.DataFrame()
for i in all_csvs:
    df_all = pd.concat([pd.read_csv(i), df_all],  join='outer' )
    df_all = df_all[df_all['NAME'].isin(LocationList)]
df_all = df_all.groupby(by='DATE').median().drop(columns=['PSUN','TSUN'], index=[2021, 2022])
```

```python
df_all
```

```python
pd.options.display.max_columns = None
df_all = df_all.dropna(axis='columns', how='any')
df_all
```
