```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import requests
import json
```

```python
# Read in Data
zillow = pd.read_csv('clean_data.csv', header=0)
crime = pd.read_csv('full_crime_data.csv', header=0)
education = pd.read_csv('education_data/Education_data.csv', header=0)
```

```python
zillow.head()
```

```python
crime = crime.rename(columns = {'Unnamed: 0':'regionidcounty'})
crime.head()
```

```python
education = education.rename(columns = {'Unnamed: 0':'regionidcounty'})
education.head()
```

```python
first_merge = pd.merge(education, crime, on='regionidcounty')
final_data = pd.merge(zillow, first_merge, on='regionidcounty')
```

```python
final_data
```

```python
longitude = final_data['longitude']
latitude = final_data['latitude']
```

```python
longitude_list = []
latitude_list = []

for i in range(0,len(final_data)):
    lat_num = (latitude[i])
    lat_num = str(lat_num)[0:2]+'.'+str(lat_num)[2:8]
    lat_num = pd.to_numeric(lat_num)
    latitude_list.append(lat_num)
    
    long_num = (longitude[i])
    long_num = str(long_num)[0:4]+'.'+str(long_num)[4:10]
    long_num = pd.to_numeric(long_num)
    longitude_list.append(long_num)
```

```python
coord_list = ""
address_list = []
startpoint = 0
endpoint = 100
MAPQUEST_API_KEY = 'XfP5gtj9agAxxwG9A6owWcjk1B74tAc2'

while endpoint <= 90200:
    for i in range(startpoint,endpoint):
        coord_list += '&location='+str(latitude_list[i])+','+str(longitude_list[i])
        
    url = ('http://www.mapquestapi.com/geocoding/v1/batch?key=' + MAPQUEST_API_KEY + coord_list)
    mapquest_req = requests.get(url)
    json_dict = json.loads(mapquest_req.text)
    json_results = json_dict['results']
    
    
    for j in range(0,100):
        address_list.append(json_results[j]['locations'][0]['street'] + ", " + json_results[j]['locations'][0]['adminArea5'] + " " +json_results[j]['locations'][0]['adminArea3'] + " " + json_results[j]['locations'][0]['postalCode'])
    
    coord_list = ""
    endpoint += 100
    startpoint += 100
```

```python
final_data['address'] = address_list
```
